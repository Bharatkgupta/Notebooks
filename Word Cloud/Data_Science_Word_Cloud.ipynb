```
# importing libraries
from wordcloud import WordCloud, ImageColorGenerator
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```
# creating word list
data = ['Data_Science', 'Big_Data', 'Deep_Learning', 'Machine_Learning', 'Data_Scientist', 'Statistics', 'Visualization', 'Python', 'R',
        'AI', 'Data_Processing', 'Data_mining', 'Clustering', 'Classification', 'Regression', 'Prediction', 'NLP', 'neural network', 'numpy',
        'scipy', 'pandas', 'scikit-learn', 'keras', 'pytorch', 'tensorflow', 'matplotlib', 'KNN', 'Gradient_Descent', 'Probabilty', 'Analysis',
        'Programming', 'Algorithm', 'Engineering', 'models', 'AR', 'VR', 'Knowledge', 'cloud_computing', 'training', 'SVM', 'decision_trees',
        'cross_validation', 'RNN', 'CNN', 'LSTM', 'distribution', 'hadoop', 'Apache_spark', 'mapreduce', 'analyst', 'dataset', 'reinforcement']
```

```
mask=np.array(Image.open("data_science.png").convert("L"))     # Mask
plt.imshow(mask)
```

```
dpic=WordCloud(width=1500, height=450, prefer_horizontal=0.7, mask=mask, background_color="white", repeat=True, max_words=8000, margin=0)  # creating world cloud instance
dpic.generate(" ".join(data))
```

```
col_img = np.array(Image.open("color_image.jpg"))
plt.imshow(col_img)
```

```
image_colors=ImageColorGenerator(col_img)  # creating color mask
dpic.recolor(color_func=image_colors)
```

```
plt.imshow(dpic)
```

```
dpic.to_file("WCD.png")
```
